```python
#DAYS EMPLOYED 外れ値の処理
df.loc[df['DAYS_EMPLOYED']>=0,'DAYS_EMPLOYED']=0
df.loc[df['DAYS_EMPLOYED']<-6000,'DAYS_EMPLOYED']=np.nan
```

```python
#DAYS PHONECHANGE 外れ値処理
df['DAYS_LAST_PHONE_CHANGE'].fillna(df['DAYS_LAST_PHONE_CHANGE'].mean(),inplace=True)
df.loc[df['DAYS_LAST_PHONE_CHANGE']<-3500,'DAYS_LAST_PHONE_CHANGE']=np.nan
```

```python
#DAYS_REGISTRATION 外れ値処理
df.loc[df['DAYS_REGISTRATION']<-15000,'DAYS_REGISTRATION']=np.nan
```

```python
#AGE
df['DAYS_BIRTH'] = df['DAYS_BIRTH'] / -365
df['DAYS_EMPLOYED'] = df['DAYS_EMPLOYED'] / -365
df['DAYS_REGISTRATION'] = df['DAYS_REGISTRATION'] / -365
df['DAYS_ID_PUBLISH'] = df['DAYS_ID_PUBLISH'] / -365
df['DAYS_LAST_PHONE_CHANGE'] = df['DAYS_LAST_PHONE_CHANGE'] / -365
```

```python
#phone
df['PHONE/BIRTH'] = df['DAYS_LAST_PHONE_CHANGE'] / df['DAYS_BIRTH']
```

```python
#同一人物を調べ

# 差分計算
df['D_REGISTRATION_BIRTH'] = df['DAYS_REGISTRATION'] - df['DAYS_BIRTH']
df['D_ID_BIRTH'] = df['DAYS_ID_PUBLISH'] - df['DAYS_BIRTH']
df['D_EMPLOYED_BIRTH'] = df['DAYS_EMPLOYED'] - df['DAYS_BIRTH']
df['D_LAST_PHONE_BIRTH'] = df['DAYS_LAST_PHONE_CHANGE'] - df['DAYS_BIRTH']

# グループ化
GROUP_F = ['D_REGISTRATION_BIRTH', 'D_ID_BIRTH', 'D_EMPLOYED_BIRTH', 'D_LAST_PHONE_BIRTH']
df['group_num'] = df.groupby(GROUP_F).ngroup()
df['group_num'].value_counts()
df.drop(['D_REGISTRATION_BIRTH','D_ID_BIRTH','D_EMPLOYED_BIRTH','D_LAST_PHONE_BIRTH'],axis=1)
```

```python
df.drop(['D_REGISTRATION_BIRTH','D_ID_BIRTH','D_EMPLOYED_BIRTH','D_LAST_PHONE_BIRTH'],axis=1,inplace=True)
```

```python
# OWN_CAR_AGEの60以上の値（外れ値）を欠損値扱いする
df.loc[df["OWN_CAR_AGE"] >= 60, "OWN_CAR_AGE"] = np.nan
# OWN_CAR_AGEをグループ分け
df["OWN_CAR_AGE"] = df["OWN_CAR_AGE"] // 10
df["OWN_CAR_AGE"].unique()
```

```python
#OBS_30_CNT 5以上の外れ値を処理
df.loc[df['OBS_30_CNT_SOCIAL_CIRCLE']>5,'OBS_30_CNT_SOCIAL_CIRCLE']=np.nan
```

```python
#Last_phone_change
df.fillna({'DAYS_LAST_PHONE_CHANGE':0},inplace=True)
df.loc[df['DAYS_LAST_PHONE_CHANGE']<=-3000,'DAYS_LAST_PHONE_CHANGE']=np.nan
df['DAYS_LAST_PHONE_CHANGE'] = df['DAYS_LAST_PHONE_CHANGE'] // 365
```

```python
#AMT_GOODS_PRICE の外れ値埋め
df.loc[df['AMT_GOODS_PRICE']>1300000,'AMT_GOODS_PRICE']=np.nan
```

```python
#AMT_INCOME_TOTAL 外れ値
df.loc[df['AMT_INCOME_TOTAL']>300000,'AMT_INCOME_TOTAL']=np.nan
```

```python
#AMT_CREDIT 外れ値
df.loc[df['AMT_CREDIT']>1600000,'AMT_CREDIT']=np.nan
```

```python
#AMT_REQ_CREDIT_BUREAU 外れ値
df.loc[df['AMT_REQ_CREDIT_BUREAU_HOUR']>=1,'AMT_REQ_CREDIT_BUREAU_HOUR']=np.nan
df.loc[df['AMT_REQ_CREDIT_BUREAU_MON']>4,'AMT_REQ_CREDIT_BUREAU_MON']=np.nan
df.loc[df['AMT_REQ_CREDIT_BUREAU_QRT']>4,'AMT_REQ_CREDIT_BUREAU_QRT']=np.nan
df.loc[df['AMT_REQ_CREDIT_BUREAU_YEAR']>7,'AMT_REQ_CREDIT_BUREAU_YEAR']=np.nan
```

```python
#欠損値を平均値で埋める
df['OWN_CAR_AGE']=df['OWN_CAR_AGE'].fillna(-9999)
df['AMT_ANNUITY'].fillna(df['AMT_ANNUITY'].mean(),inplace=True)
df['AMT_GOODS_PRICE'].fillna(df['AMT_GOODS_PRICE'].mean(),inplace=True)
df['CNT_FAM_MEMBERS'].fillna(df['CNT_FAM_MEMBERS'].mean(),inplace=True)
```

```python
#男女をdecimelnum
df['CODE_GENDER'] = df['CODE_GENDER'].replace({"F":0,"M":1,"XNA":-9999})
```

```python
#CONTRACT_TYPE
df['NAME_CONTRACT_TYPE'] = df['NAME_CONTRACT_TYPE'].replace({"Cash loans":0,"Revolving loans":1})
```

```python
#EXRを職業別groupbyのmean
df['EXT_SOURCE_1'] = df.groupby('ORGANIZATION_TYPE')['EXT_SOURCE_1'].transform(lambda x: x.fillna(x.mean()))
df['EXT_SOURCE_2'] = df.groupby('ORGANIZATION_TYPE')['EXT_SOURCE_2'].transform(lambda x: x.fillna(x.mean()))
df['EXT_SOURCE_3'] = df.groupby('ORGANIZATION_TYPE')['EXT_SOURCE_3'].transform(lambda x: x.fillna(x.mean()))
```

```python
#FAMILY AND CHILDREN
#3人以上子供を持つ人をまとめる
df.loc[df['CNT_CHILDREN']>=3,'CNT_CHILDREN']=3
#家族が５人以上をまとめる
df.loc[df['CNT_FAM_MEMBERS']>=5,'CNT_FAM_MEMBERS']=5
#家族の中の子供の割合を求める
df['CNT_CHILDREN_RATIO']=df['CNT_CHILDREN']/df['CNT_FAM_MEMBERS']
df.drop(['CNT_CHILDREN'],axis=1,inplace=True)
```

```python
#特徴量
#EXT 四則演算
df['EXT_123_mean'] = (df['EXT_SOURCE_1'] + df['EXT_SOURCE_2'] + df['EXT_SOURCE_3']) / 3
df['EXT_23_mean'] = (df['EXT_SOURCE_2'] + df['EXT_SOURCE_3']) / 2
df['EXT_12_mean'] = (df['EXT_SOURCE_1'] + df['EXT_SOURCE_2']) / 2
df['EXT_13_mean'] = (df['EXT_SOURCE_1'] + df['EXT_SOURCE_3']) / 2
df['EXT_23_sabun'] = abs(df['EXT_SOURCE_2'] - df['EXT_SOURCE_3'])
df['EXT_12_sabun'] = abs(df['EXT_SOURCE_1'] - df['EXT_SOURCE_2'])
df['EXT_13_sabun'] = abs(df['EXT_SOURCE_1'] - df['EXT_SOURCE_3'])
df['EXT_123_*']=df['EXT_SOURCE_1']*df['EXT_SOURCE_2']*df['EXT_SOURCE_3']

#AMT

df['AMT_HOUR_YEAR']=df['AMT_REQ_CREDIT_BUREAU_HOUR']+df['AMT_REQ_CREDIT_BUREAU_MON']+df['AMT_REQ_CREDIT_BUREAU_QRT']+df['AMT_REQ_CREDIT_BUREAU_YEAR']
df['AMT_mean']=(df['AMT_ANNUITY']+df['AMT_CREDIT']+df['AMT_GOODS_PRICE'])/3


#OBS+DEF
df['OBS_DEF_30_mean']=(df['OBS_30_CNT_SOCIAL_CIRCLE']+df['DEF_30_CNT_SOCIAL_CIRCLE'])/2



#AMT 処理
df['CREDIT_ANNUITY'] = df['AMT_ANNUITY'] / df['AMT_ANNUITY']
df['CREDIT_GOODS_PRICE'] = df['AMT_CREDIT'] / df['AMT_GOODS_PRICE']
df['INCOME_TOTAL_ANNUITY'] = df['AMT_ANNUITY'] / df['AMT_INCOME_TOTAL']
df['INCOME_TOTAL_CREDIT'] = df['AMT_INCOME_TOTAL'] / df['AMT_CREDIT']

df['AMT_GOOODS-CREDIT'] = df['AMT_GOODS_PRICE'] - df['AMT_CREDIT']

df['DAYS_EMPLOYED_/'] = df['DAYS_EMPLOYED'] / df['DAYS_BIRTH']

#FLAG*age
df['FLAG_EMAIL_PHONE_MOBILE']= df['FLAG_EMAIL'] + df['FLAG_MOBIL'] + df['FLAG_PHONE']+ df['FLAG_EMP_PHONE'] + df['FLAG_WORK_PHONE'] + df['FLAG_CONT_MOBILE']
df['Reliability']= df['DAYS_BIRTH'] + df['CODE_GENDER'] + df['DAYS_EMPLOYED'] + df['FLAG_EMAIL_PHONE_MOBILE']+df['CNT_CHILDREN_RATIO']
#使ったカラムを消去
df.drop(['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3'],axis=1,inplace=True)
df.drop(['AMT_ANNUITY','AMT_CREDIT','AMT_GOODS_PRICE','DAYS_BIRTH','DAYS_EMPLOYED'],axis=1,inplace=True)
df.drop(['FLAG_EMAIL_PHONE_MOBILE','FLAG_EMAIL','FLAG_MOBIL','FLAG_PHONE','FLAG_EMP_PHONE','FLAG_WORK_PHONE','FLAG_CONT_MOBILE'],axis=1,inplace=True)
df.drop(['FLAG_OWN_REALTY','DAYS_LAST_PHONE_CHANGE','CREDIT_ANNUITY','INCOME_TOTAL_CREDIT'],axis=1,inplace=True)
```

```python
#LightGBM
from gc import collect
from lightgbm import LGBMClassifier as LGBMC
import optuna
from optuna import Trial, trial, create_study
from sklearn.model_selection import cross_val_score
```

```python
def objective(trial):
    clf = LGBMC(max_depth  = trial.suggest_int('max_depth',     2,  5),        # 決定木の深さ
                subsample  = trial.suggest_float('subsample',   0,   1),        # 決定木ごとに学習データの行をサンプリングする割合
                colsample_bytree =\
                             trial.suggest_float('colsample_bytree',   0.05,   1), # 決定木ごとに特徴量の列をサンプリングする割合
                num_leaves = trial.suggest_int('num_leaves',   40, 200),
                reg_alpha  = trial.suggest_float('reg_alpha',   1e-2,   1),        # 決定木の葉のウェイトに対するL1正則化の強さ
                reg_lambda = trial.suggest_float('reg_lambda',  1e-2,   1),          # 決定木の葉のウェイトに対するL2正則化の強さ
                )
    score = cross_val_score(clf, X_train, y_train, cv = 3)
    accuracy = score.mean()
    return accuracy

# Study(トライアルを繰り返す一連の最適化プロセス)オブジェクトの呼び出し
study = optuna.create_study(direction = 'maximize')

# 最適化のループを呼び出す
study.optimize(objective, n_trials = 100)

print()
collect()
print(f'Best objective value: {study.best_value}')
print(f'Best parameter: {study.best_params}')
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import KFold
from lightgbm import LGBMClassifier, early_stopping, log_evaluation
from sklearn.metrics import mean_absolute_error
import numpy as np
import pandas as pd

# ベストパラメータの設定
best_params = {
    'max_depth': 5,
    'subsample': 0.5214315801643437,
    'colsample_bytree': 0.3611259239991222,
    'num_leaves': 100,
    'reg_alpha': 0.9765132171763462,
    'reg_lambda': 0.8694429902758802
}

# skf の設定
skf = StratifiedKFold(n_splits=8, shuffle=True, random_state=42)

# 評価結果とモデルを保存するリスト
scores = []
models = []

# K-Foldでのモデルの訓練と評価
for fold, (train_indices, valid_indices) in enumerate(skf.split(X,y)):
    X_train, X_valid = X[train_indices], X[valid_indices]
    y_train, y_valid = y[train_indices], y[valid_indices]

    # モデルの初期化
    model = LGBMClassifier(
        max_depth=best_params['max_depth'],
        subsample=best_params['subsample'],
        colsample_bytree=best_params['colsample_bytree'],
        num_leaves=best_params['num_leaves'],
        reg_alpha=best_params['reg_alpha'],
        reg_lambda=best_params['reg_lambda'],
        learning_rate=0.01,
        n_estimators=1000,
        random_state=42
    )

    # モデルの学習
    model.fit(
        X_train, y_train,
        eval_set=[(X_valid, y_valid)],
        eval_metric='logloss',
        callbacks=[
            early_stopping(stopping_rounds=50),
            log_evaluation(period=100)
        ]
    )

    # モデルを保存
    models.append(model)

    # 検証データでのスコア計算

    lgb_train_pred = model.predict_proba(X_train)[:, 1]
    lgb_valid_pred = model.predict_proba(X_valid)[:, 1]
    y_valid_pred = model.predict(X_valid)
    score = mean_absolute_error(y_valid, y_valid_pred)
    scores.append(score)
    print(f"Fold MAE: {score:.4f}")

# 平均スコアの計算
average_score = np.mean(scores)
print(f"K-Fold の平均 MAE スコア: {average_score:.4f}")

print(f"Train Score: {roc_auc_score(y_train, lgb_train_pred)}")
print(f"Valid Score: {roc_auc_score(y_valid, lgb_valid_pred)}")
```
